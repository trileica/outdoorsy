# Data Imputation

```python
# check n/a
df_hour.isnull().sum()
```

```python
df_hour[df_hour['registered'].isnull()]
```

```python
df_hour[df_hour['casual'].isnull()]
```

```python
# fill na with 0 because casual + registered = cnt
df_hour['casual'].fillna(0, inplace = True)
df_hour['registered'].fillna(0, inplace = True)
df_hour.isnull().sum()
```

```python
df_hour[df_hour['holiday'].isnull()]
```

```python
# fill na with 1 because 1/1 is holiday and only 1 record is missing
df_hour['holiday'].fillna(1, inplace = True)
df_hour['holiday'] = df_hour['holiday'].astype(int)
```

```python
df_hour[df_hour['atemp'].isnull()]
```

## Multivariate imputation from weather related columns

```python
# read file just for plotting
df_hour_ = pd.read_csv(FILE_NAME)
```

```python
df_hour_.plot(x = 'dteday', y = 'atemp', figsize = (20, 5))
```

```python
df_hour_.plot(x = 'dteday', y = 'hum', figsize = (20, 5))
```

```python
df_hour_.plot(x = 'dteday', y = 'windspeed', figsize = (20, 5))
```

```python
# because those 3 columns are weather related, it makes more sense to use their neighbors to impute their values
imp = IterativeImputer(max_iter = 10, random_state = 0)
imp.fit(df_hour[['temp', 'atemp', 'hum', 'windspeed', 'weathersit']])
na_index = df_hour[['atemp', 'hum', 'windspeed']][df_hour.isnull().any(axis = 1)].index

imp_to_fit = df_hour.loc[na_index][['temp', 'atemp', 'hum', 'windspeed', 'weathersit']]
imp_fit = imp.transform(imp_to_fit)

df_imp = pd.concat([pd.DataFrame(imp_fit), pd.DataFrame({'col' : na_index})], axis = 1)
df_imp.columns = ['temp', 'atemp', 'hum', 'windspeed', 'weathersit', 'index']
df_imp.set_index('index', inplace = True)
df_hour = df_hour.merge(df_imp[['atemp', 'hum', 'windspeed']], right_index = True, left_index = True, how = 'left')
```

```python
df_hour['atemp'] = df_hour.apply(lambda x : x['atemp_x'] if x['atemp_x'] == x['atemp_x'] else x['atemp_y'], axis = 1)
df_hour['hum'] = df_hour.apply(lambda x : x['hum_x'] if x['hum_x'] == x['hum_x'] else x['hum_y'], axis = 1)
df_hour['windspeed'] = df_hour.apply(lambda x : x['windspeed_x'] if x['windspeed_x'] == x['windspeed_x'] else x['windspeed_y'], axis = 1)
```

```python
# check again to make sure no na
df_hour.drop(columns = [i for i in df_hour.columns if "_" in i], inplace = True)
df_hour.isnull().sum()
```

# Feature Engineering

```python
df_hour_dummy = pd.concat([df_hour,                     
                     pd.get_dummies(df_hour['season'], prefix = 'season'),
                     pd.get_dummies(df_hour['yr'], prefix = 'yr'),
                     pd.get_dummies(df_hour['mnth'], prefix = 'mnth'),
                     pd.get_dummies(df_hour['hr'], prefix = 'hr'),
                     pd.get_dummies(df_hour['holiday'], prefix = 'holiday'),
                     pd.get_dummies(df_hour['weekday'], prefix = 'weekday'),
                     pd.get_dummies(df_hour['workingday'], prefix = 'workingday'),
                     pd.get_dummies(df_hour['weathersit'], prefix = 'weathersit'),
                    ],
                    axis = 1)
```

# Preparation For Training/Testing

```python
COL_Y = 'cnt'
COL_TO_DROP = ['casual', 'registered'] # take out columns that might be data leaking

ys = df_hour[COL_Y]
xs = df_hour.drop(columns = [COL_Y] + COL_TO_DROP)
X_train, X_test, y_train, y_test = train_test_split(xs, ys,
                                                    test_size = 0.1, random_state = 42)

X_train.columns
```

# Model Training With a Few Selections

```python
models = [RandomForestRegressor(), AdaBoostRegressor(), BaggingRegressor(), KNeighborsRegressor(), SVR()]
model_names = ['Random Forest', 'AdaBoost', 'Bagging', 'KNeighborsRegressor', "SVR"]

for i in range(len(models)):    
    clf = models[i]
    time_pre = time.time()
    clf.fit(X_train, y_train)
    time_post = time.time()
    test_pred = clf.predict(X_test)
    print(f"Algorithm : {model_names[i]}")
    print(f"  RMSE : {round(np.sqrt(mean_squared_error(test_pred, y_test)), 3)}")
    print(f"  Total time : {round(time_post - time_pre, 3)}")
    print("")
    
```

```python
cat_columns = [0, 1, 2, 3, 4, 5, 6, 7, 9, 10]

clf = CatBoostRegressor
train_pool = Pool(X_train, y_train, 
                  cat_features = cat_columns)
test_pool = Pool(X_test, 
                 cat_features = cat_columns)

cat_clf = CatBoostRegressor(iterations = 1000, 
                          depth = 10, 
                          learning_rate = 0.1, 
                          loss_function = 'RMSE')

time_pre = time.time()
cat_clf.fit(train_pool, verbose = False)
time_post = time.time()

preds = cat_clf.predict(test_pool)
print(f"Algorithm : CatBoost")
print(f"  RMSE : {round(np.sqrt(mean_squared_error(preds, y_test)), 3)}")
print(f"  Total time : {round(time_post - time_pre, 3)}")
```

```python
COL_TO_DROP = ['season', 'day', 'yr', 'mnth', 'hr', 'holiday', 'weekday', 'workingday', 'weathersit', 'year']
ys_d = df_hour_dummy[COL_Y]
xs_d = df_hour_dummy.drop(columns = [COL_Y] + COL_TO_DROP)

X_train_d, X_test_d, y_train_d, y_test_d = train_test_split(xs_d, ys_d,
                                                            test_size = 0.1, random_state = 42)
xgb_clf = XGBRegressor()

time_pre = time.time()
xgb_clf.fit(X_train_d, y_train_d)
time_post = time.time()

preds = xgb_clf.predict(X_test_d)
print(f"Algorithm : XGBoost")
print(f"  RMSE : {round(np.sqrt(mean_squared_error(preds, y_test_d)), 3)}")
print(f"  Total time : {round(time_post - time_pre, 3)}")
```

# Search For Best Parameters


```python
xgb_model = XGBRegressor()

param_grid = {
        'max_depth': [4, 6, 8, 10, 20],
        'min_child_weight': [0.1, 1, 2],
        'learning_rate': [0.001, 0.01, 0.1, 0.3, 0.5, 1],
}

kfold = KFold(n_splits = 10, shuffle = True, random_state = 10)
grid_search = RandomizedSearchCV(xgb_model, param_grid, scoring = "neg_mean_squared_error", n_iter = 100, cv = kfold)
grid_result = grid_search.fit(X_train_d, y_train_d)

# summarize results
print("Best: %f using %s" % (grid_result.best_score_, grid_result.best_params_))
means = grid_result.cv_results_[ 'mean_test_score' ]
stds = grid_result.cv_results_[ 'std_test_score' ]
params = grid_result.cv_results_[ 'params' ]
```
